```python
import os 
import AwsClient as aws
import requests
```

```python
s3_client = aws.s3_client(
    aws_account_access_key = os.getenv('AWS_ACCESS_KEY_ID'),
    aws_secret_account_access_key = os.getenv('AWS_SECRET_ACCESS_KEY'),
    region = 'us-east-1'
)
```

```python
r = requests.get('https://financialmodelingprep.com/stable/historical-chart/5min?symbol=AAPL&apikey=370b0e6c12a3eb14463c34c6277abec3')
response = r.json()
```

```python
aws.json_upload_to_s3(
    instance = s3_client,
    data = response,
    bucket_name = os.getenv('S3_BUCKET_NAME'),
    file_path = 'test/test/test',
    file_name='appl-5-min-s3-test'
)
```

```python
data = aws.json_get_from_s3(
    instance = s3_client,
    bucket_name = os.getenv('S3_BUCKET_NAME'),
    file_path = 'test/test/test',
    file_name = 'appl-5-min-s3-test'
)
data
```
